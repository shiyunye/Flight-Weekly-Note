```python

def format_number(num):
    if pd.isna(num):
        return ''
    if abs(num) >= 1e6:
        return f"{num/1e6:.1f}M"
    elif abs(num) >= 1e3:
        return f"{num/1e3:.0f}K"
    elif abs(num) < 1e3:
        return "<1K"
    return f"{num:.0f}"

def format_percentage(num):
    if pd.isna(num):
        return ''
    return f"{num:.1f}%"

def round_to_nearest_10(num):
  return round(num / 10) * 10

df_pricelince=df_weekly[(df_weekly['brand']== 'Priceline')]
df_pricelince_b2c=df_weekly[(df_weekly['brand']== 'Priceline')&(df_weekly['company']== 'Priceline B2C')]
df_pricelince_b2c_standalone=df_weekly[(df_weekly['brand']== 'Priceline')&(df_weekly['company']== 'Priceline B2C')&(df_weekly['product']== 'Flights Only')]
```

```python
# Create df_mor
df_mor= pd.DataFrame()
df_mor['Net Contr'] = df_pricelince[(df_pricelince['wk_ending'] == end_date)& ((df_pricelince['merchant_of_record'] == 'VCC')|(df_pricelince['merchant_of_record'] == 'PCLN'))].groupby(['offer_method_code'])['net_contribution_cy'].sum().round(-1).astype(int)
df_mor['Net Contr_cwly'] = df_pricelince[(df_pricelince['wk_ending'] == end_date)& ((df_pricelince['merchant_of_record'] == 'VCC')|(df_pricelince['merchant_of_record'] == 'PCLN'))].groupby(['offer_method_code'])['net_contribution_ly'].sum().round(-1).astype(int)
df_mor['Net Contr_PW'] = df_pricelince[(df_pricelince['wk_ending'] == pp_end_date)& ((df_pricelince['merchant_of_record'] == 'VCC')|(df_pricelince['merchant_of_record'] == 'PCLN'))].groupby(['offer_method_code'])['net_contribution_cy'].sum().round(-1).astype(int)
df_mor['Net Contr_PWly'] = df_pricelince[(df_pricelince['wk_ending'] == pp_end_date)& ((df_pricelince['merchant_of_record'] == 'VCC')|(df_pricelince['merchant_of_record'] == 'PCLN'))].groupby(['offer_method_code'])['net_contribution_ly'].sum().round(-1).astype(int)


# Calculate YoY and YoY PW
df_mor.loc['Total','Net Contr':'Net Contr_PWly']=[df_mor['Net Contr'].sum(),df_mor['Net Contr_cwly'].sum(),df_mor['Net Contr_PW'].sum(), df_mor['Net Contr_PWly'].sum()]


# Create df_mor_1
df_mor_1= pd.DataFrame()
df_mor_1['Net Contr'] = df_pricelince[(df_pricelince['wk_ending'] == end_date)].groupby(['offer_method_code'])['net_contribution_cy'].sum().round(-1).astype(int)
df_mor_1['Net Contr_cwly'] = df_pricelince[(df_pricelince['wk_ending'] == end_date)].groupby(['offer_method_code'])['net_contribution_ly'].sum().round(-1).astype(int)
df_mor_1['Net Contr_PW'] = df_pricelince[(df_pricelince['wk_ending'] == pp_end_date)].groupby(['offer_method_code'])['net_contribution_cy'].sum().round(-1).astype(int)
df_mor_1['Net Contr_PWly'] = df_pricelince[(df_pricelince['wk_ending'] == pp_end_date)].groupby(['offer_method_code'])['net_contribution_ly'].sum().round(-1).astype(int)
# Calculate YoY and YoY PW
df_mor_1.loc['Total','Net Contr':'Net Contr_PWly']=[df_mor_1['Net Contr'].sum(),df_mor_1['Net Contr_cwly'].sum(),df_mor_1['Net Contr_PW'].sum(), df_mor_1['Net Contr_PWly'].sum()]

df_mor = pd.concat([df_mor,df_mor_1.add_suffix('_total')], axis=1)


df_mor['Actual']=(df_mor['Net Contr']/df_mor['Net Contr_total']*100).round(1).apply(format_percentage)
df_mor['%mor_ly']=(df_mor['Net Contr_cwly']/df_mor['Net Contr_cwly_total']*100).round().apply(format_percentage)
df_mor['YoY_bps']=(((df_mor['Net Contr']/df_mor['Net Contr_total']*100)-(df_mor['Net Contr_cwly']/df_mor['Net Contr_cwly_total']*100))*100).round()
df_mor['Actual_PW']=(df_mor['Net Contr_PW']/df_mor['Net Contr_PW_total']*100).round().apply(format_percentage)
df_mor['YoY_PW_bps']=(((df_mor['Net Contr_PW']/df_mor['Net Contr_PW_total']*100)-(df_mor['Net Contr_PWly']/df_mor['Net Contr_PWly_total']*100))*100).round()
df_mor['YoY_bps']=round_to_nearest_10(df_mor['YoY_bps'])
df_mor['YoY_PW_bps']=round_to_nearest_10(df_mor['YoY_PW_bps'])
df_mor=df_mor.loc[['Retail (Disclosed)', 'Total'],['Actual','YoY_bps','YoY_PW_bps']]
df_mor
```
